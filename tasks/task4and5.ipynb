```python
file_path = "/content/drive/MyDrive/my_text_file.txt"
with open(file_path, "w") as f:
  f.write("Это мой тестовый файл в Google Drive\n")
  f.write("Это мой любимый тестовый файл в Google Drive\n")
print("Файл сохранен в Google Drive")
with open(file_path, "r") as f:
  content = f.read()
  print("Содержимое файла: \n", content)
```

```python
students = [
    ["Иванов И.И.", 20, "ИВТ-101"],
    ["Петров П.П.", 22, "ИВТ-102"],
    ["Сидорова А.А.", 21, "ИВТ-103"],
]

csv_path = "/content/drive/MyDrive/students.csv"
with open(csv_path, "w") as f:
  for student in students:
    f.write(",".join(map(str, student)) + "\n")

print("Файл students.csv  сохранен в Google Drive.")
```

```python
from google.colab import files
uploaded = files.upload()
import csv
input_file = "students.csv"
with open(input_file, "r", encoding="utf-8") as f:
  reader = csv.reader(f)
  data = list(reader)
print("Содержимое файла:")
for row in data:
  print(row)
```

```python
data.append(["Коробка В.А.", 19, "ИВТ-6-о-24-1"])
for row in data:
  print(row)
output_path = "students.csv"
with open(output_path, "w", newline="", encoding="utf-8") as f:
  writer = csv.writer(f)
  writer.writerows(data)
print("Файл сохранён обратно в students.csv!")
```
