## Определение обратной матрицы

Матрица $A^{-1}$ называется обратной к квадратной матрице $A$, если
$$A \cdot A^{-1} = A^{-1} \cdot A = E$$
где $E$ — единичная матрица.
Обратная матрица существует, если
$$\det(A) \neq 0.$$

```python
import os

with open("task2.txt", "w") as f:
    f.write("Один\n")
    f.write("Два\n")
    f.write("Три\n")
    f.write("Четыре\n")

with open("task2.txt", "r") as f:
    content = f.read()
    print("Содержимое файла:\n", content)

print("Файл существует:", os.path.exists("task2.txt"))

os.remove("task2.txt")
print("Файл удален!")
```
